# Array Indexing #

```python
import numpy as np
```

### 1. Numpy indexing and slicing of vectors ###

```python
arr = np.arange(11,21)
arr
```

```python
arr[6]  # indexing
```

```python
arr[1:5]  # slicing
```

```python
arr[1:5:2]  # slicing with step
```

```python
arr[:6]  # Print index 0 to 5
```

```python
arr[3:]  # print index 3 to n
```

```python
arr_slice = arr[:]  # print 0 to n
arr_slice
```

```python
arr_slice = arr[::]
arr_slice
```

```python
arr_slice = arr[3::2]
arr_slice
```

### 2. Numpy indexing and slicing of matrix ###

```python
arr = np.arange(1,31).reshape(6,5)
arr  
```

```python
arr[1]  # print row
```

```python
arr[3,2]     # indexing
```

```python
slice = arr[0:2,1:3]    #slicing
slice
```

```python
slice1 = arr[3:,3:]  # slicing
slice1
```

```python
arr[:,2]  #print column
```

### 3. Boolean indexing ###

```python
arr = np.arange(11,21)
arr
```

```python
bool_index = arr%2==0 
bool_index
```

```python
arr = arr[bool_index]
arr
```

```python
import numpy as np

# columns : [Age, Math_Marks, Science_Marks, English_Marks]
data = np.array([
    [15, 85, 90, 80],    # Student 1
    [16, 78, 88, 82],    # Student 2
    [15, 92, 95, 88],    # Student 3
    [17, 80, 85, 78],    # Student 4
    [16, 88, 91, 85]     # Student 5
    ])
```

```python
#Get the shape of the matrix
data.shape
```

```python
# find the average age of students
np.mean(data[:,0])
```

```python
# extract math marks of all the students
data[:,1]
```

```python
# find the higest science marks
np.max(data[:,2])
```

```python
# get the details of the student who scored more than 90 marks in maths
data[data[:,1] > 90]
```

```python
# increase all students' math marks by 5
data[:,1] += 5
data
```

```python
# find how many students younger than 16 years are there
len(data[data[:,0] < 16])
```

```python
# calculate the average marks in each subject(column-wise average)
np.mean(data[:,1:], axis=0)
```

```python
# get the data of the student who scored at least 80 in both maths and science
data[(data[:,1] >= 80) & (data[:,2] >= 80)]
```

```python
# replace all the science marks less than 75 replace with 0
data[data[:,2] < 90, 2] = 0
data
```

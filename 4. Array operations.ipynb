### 3. Deep and Shallow Copy ###

```python
a = np.arange(1,21)
a
```

```python
slice = a[:5]
slice = slice*10  # it is a shallow copy, the original array is affected by changes to the slice.
# The original array is modified after changing the slice, demonstrating that it is a shallow copy.
```

```python
a[:2]*10
a        # in deep copy, the original array is not affected by changes to the slice.
# The original array remains unchanged after modifying the slice, demonstrating that it is a deep copy.
```

```python
b = a 
b[0] = 100
a  # in assignment, the original array is affected by changes to the new variable, demonstrating that it is a shallow copy.
```

### 5. Advance Array Manipulation ###

#####  *Stacking Array* #####

```python
# we cannot stack two matrixs or we stack two vectors only
a = np.array([1,2,3,4])
b = np.array([5,6,7,8])
```

```python
np.vstack((a,b))  # vertical stacking of arrays
# vertical stack is used to stack arrays vertically (row-wise).
```

```python
np.hstack((a,b))  # horizontal stacking of arrays
# horizontal stack is used to stack arrays horizontally (column-wise).
```

```python
np.column_stack((a,b))  # column stacking of arrays
# column stack is used to stack 1-D arrays as columns into a 2-D array.
# It is similar to hstack but it treats 1-D arrays as columns rather than rows.
```
